```python
#Write code

from nltk.util import ngrams
def extract_ngrams(data, num):
    n_grams = ngrams(data, num)
    return [' '.join(grams) for grams in n_grams]

unigrams=[]
bigrams=[]
trigrams=[]
fourgrams=[] 
unigrams = extract_ngrams(words,1) 
bigrams = extract_ngrams(words,2) 
trigrams = extract_ngrams(words,3) 
fourgrams = extract_ngrams(words,4)
    ##similar for trigrams and fourgrams
```

```python
#stopwords = code for downloading stop words through nltk
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english')) 
#print top 10 unigrams, bigrams after removing stopwords
uni_processed = [p for p in unigrams if p not in stopwords]
fdist = nltk.FreqDist(uni_processed)
fdist1 = dict(fdist)
fdist1 = [(v,k) for k, v in fdist1.items()]
fdist1 = sorted(fdist1,reverse=True) 
print("The top 10 unigrams are as follows :") 
for i in range(10):
    print(fdist1[i][1])
print("\n")
#print top 10 bigrams, trigrams, fourgrams after removing stopwords
bi_processed=[]
for p in bigrams:
    count=0
    word=word_tokenize(p)
    for w in word:
        if w in stopwords:
            count=count+1
    if(count<2):
        bi_processed.append(p)
fdist = nltk.FreqDist(bi_processed)
fdist2 = dict(fdist)
fdist2 = [(v,k) for k, v in fdist2.items()]
fdist2 = sorted(fdist2,reverse=True) 
print("The top 10 bigrams are as follows :") 
for i in range(10):
    print(fdist2[i][1]+" "+str(fdist2[i][0]))
print("\n")
tri_processed=[]
for p in trigrams:
    count=0
    word=word_tokenize(p)
    for w in word:
        if w in stopwords:
            count=count+1
    if(count<3):
        tri_processed.append(p)
fdist = nltk.FreqDist(tri_processed)
fdist3 = dict(fdist)
fdist3 = [(v,k) for k, v in fdist3.items()]
fdist3 = sorted(fdist3,reverse=True) 
print("The top 10 trigrams are as follows :") 
for i in range(10):
    print(fdist3[i][1]+" "+str(fdist3[i][0]))
print("\n")
four_processed=[]
for p in fourgrams:
    count=0
    word=word_tokenize(p)
    for w in word:
        if w in stopwords:
            count=count+1
    if(count<4):
        four_processed.append(p)
fdist = nltk.FreqDist(four_processed)
fdist4 = dict(fdist)
fdist4 = [(v,k) for k, v in fdist4.items()]
fdist4 = sorted(fdist4,reverse=True) 
print("The top 10 fourgrams are as follows :") 
for i in range(10):
    print(fdist4[i][1]+" "+str(fdist4[i][0]))
```

# Applying Smoothing


Assume additional training data in which each possible N-gram occurs exactly once and adjust estimates.

> ### $ Probability(ngram) = \frac{Count(ngram)+1}{ N\, +\, V} $

N: Total number of N-grams <br>
V: Number of unique N-grams



```python
from math import *
#You are to perform Add-1 smoothing here:
v=0
uni_unique=[]
for i in unigrams:
    if i not in uni_unique:
        uni_unique.append(i)
        v=v+1
fdist = nltk.FreqDist(unigrams)
fdist = dict(fdist)
fdist = [(l,k) for k,l in fdist.items()]
n=len(fdist)
sum=0
for i in range(n):
    sum=sum+fdist[i][0]
uni_smoothed=[]
for i in range(n):
    a=(fdist[i][0]+1)/(sum+v)
    uni_smoothed.append((fdist[i][1],a))
uni_smoothed.sort(key = lambda x: x[1],reverse=True)
print("The top 10 unigrams are as follows :") 
for i in range(10):
    print(uni_smoothed[i][0]+" "+str(uni_smoothed[i][1]))
print("\n")
#write similar code for bigram, trigram and fourgrams
fdist = nltk.FreqDist(bigrams)
fdist = dict(fdist)
fdist = [(l,k) for k,l in fdist.items()]
n=len(fdist)
sum=0
for i in range(n):
    sum=sum+fdist[i][0]
bi_smoothed=[]
for i in range(n):
    a=(fdist[i][0]+1)/(sum+pow(v,2))
    bi_smoothed.append((fdist[i][1],a))
bi_smoothed.sort(key = lambda x: x[1],reverse=True)
print("The top 10 bigrams are as follows :") 
for i in range(10):
    print(bi_smoothed[i][0]+" "+str(bi_smoothed[i][1]))
print("\n")
#Print top 10 unigram, bigram, trigram, fourgram after smoothing
fdist = nltk.FreqDist(bigrams)
fdist = dict(fdist)
fdist = [(l,k) for k,l in fdist.items()]
n=len(fdist)
sum=0
for i in range(n):
    sum=sum+fdist[i][0]
tri_smoothed=[]
for i in range(n):
    a=(fdist[i][0]+1)/(sum+pow(v,3))
    tri_smoothed.append((fdist[i][1],a))
tri_smoothed.sort(key = lambda x: x[1],reverse=True)
print("The top 10 trigrams are as follows :") 
for i in range(10):
    print(tri_smoothed[i][0]+" "+str(tri_smoothed[i][1]))
print("\n")
fdist = nltk.FreqDist(bigrams)
fdist = dict(fdist)
fdist = [(l,k) for k,l in fdist.items()]
n=len(fdist)
sum=0
for i in range(n):
    sum=sum+fdist[i][0]
four_smoothed=[]
for i in range(n):
    a=(fdist[i][0]+1)/(sum+pow(v,4))
    four_smoothed.append((fdist[i][1],a))
four_smoothed.sort(key = lambda x: x[1],reverse=True)
print("The top 10 fourgrams are as follows :") 
for i in range(10):
    print(four_smoothed[i][0]+" "+str(four_smoothed[i][1]))
print("\n")
```
